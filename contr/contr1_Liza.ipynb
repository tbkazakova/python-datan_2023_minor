# Контрольная работа

### Что делать?

Человек записывает себе напоминалки, что надо сделать. Он может несколько раз вспомнить про одно и то же, но делать это надо только один раз.

Принимайте от пользователя строки с делами, пока он не введёт пустую строку.

После завершения выведите все дела по одному разу, игнорируя регистр (`ДЗ` и `Дз` - одно и то же дело) и отсортировав по длине последовательности (`ДЗ` короче, чем `проект`). Сообщите, сколько всего получилось.

Пример:

ВВОД:
```
Дз
смонтировать фильм
ДЗ
порядок
проект
ВДНХ, концерт (не забыть конфеты)
проверить кр
ДЗ
```
ВЫВОД:
```
дз
проект
порядок
проверить кр
смонтировать фильм
вднх, концерт (не забыть конфеты)
Всего: 6
```
