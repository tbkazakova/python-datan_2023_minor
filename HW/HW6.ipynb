## 1. Покажите фокус (2-3 балла)

Попробуйте себя в роли иллюзиониста и покажите, как ловко жонглируете кучей чисел.

Зрители предлагают вам числа.
Произведите разные действия вроде сложения, умножения, вычитания, деления, возведения в степень и так далее.
Комментируйте эти действия и в итоге получите обратно исходные числа.

БАЛЛЫ:  
- 2 балла за повторение примера ниже
- \+ бонусный балл (3 в итоге) за свой алгоритм с 3-4 действиями, который из массива исходных чисел через несколько этапов получает вновь этот же массив

ФОРМАТ ВВОДА:  
Пользователь вводит несколько целых чисел, через пробел.

ФОРМАТ ВЫВОДА:  
Программа печатает действие, которое будет выполнено с каждым исходным числом, и результат этих действий: числа через пробел.  
Последнее действие должно привести к исходным числам.


ПРИМЕР:

    1 4 5 10 12 67 3489

    Прибавлю 4!
     5 8 9 14 16 71 3493
    Умножу на 8!
     40 64 72 112 128 568 27944

     12 24 28 48 56 276 13964
    А теперь вычту 8
