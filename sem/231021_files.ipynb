## Кодировка

Сперва скажем пару слов о **кодировках**.
    Написанный текст сохраняется в памяти компьютера в виде кода: каждому символу соответствует его кодовый номер.
    А при открытии все кодовые номера снова превращаются в символы.
    Правила кодирования, то есть соответствия кодового номера и символа, могут быть разными.
    Наверняка вам встречались текстовые документы, которые при открытии оказывались не текстом, а набором странных значков.
    Это происходило потому, что при открытии файла использовалась не та кодировка, которая была при сохранении.
    Мы будем работать с файлами в кодировке UTF-8 (https://ru.wikipedia.org/wiki/UTF-8).
    Ее используют при работе с кириллицей и другими европейскими языками.

Итог:
- не бывает "просто текста", каждый текст как-то да закодирован.

## С какими файлами мы будем работать

### - файлы с расширением .txt — текстовые файлы
Их можно открыть в любом текстовом редакторе (например, Блокноте). В конце таких файлов лучше оставлять перенос строки, чтобы последней была пустая строка. Главное, чтобы сохранялась структура строк и столбцов. Тогда эти файлы можно будет открыть не только текстовом редакторе (например, Блокноте), но и в редакторе таблиц (например, Excel и Google Sheets) — чтобы получить классическое табличное представление. ## Открываем .txt

**План чтения файла:**
1. найти этот файл на компьютере
2. открыть
3. взять из него все, что нужно
4. закрыть файл

 _нахождение файла на компьютере_:  
  Чтобы программа смогла найти файл по названию, проще всего **положить его в ту же папку, где лежит файл с кодом программы**.
  Иначе придется прописывать путь к файлу ([подробнее о пути к файлу здесь](https://desktop.arcgis.com/ru/arcmap/10.3/tools/supplement/pathnames-explained-absolute-relative-unc-and-url.htm)).
    Если программа не найдет файл с таким именем, будет выдана ошибка: _файл не найден_.  

````python
# Открываем файл: пишем название файла и выбираем ему псевдоним
with open('first_text.txt') as f:
    print(f)  # Пробуем распечатать считанный файл
````

При попытке распечатать считанный файл получаем следующую информацию:

* `_io.TextIOWrapper` — это служебная информация о том, как Python внутри себя видит файл;
* `name='first_text.txt'` показывает имя исходного текстового файла, для которого мы выбрали псевдоним `infile`;
* `mode='r'` означает, что мы открыли этот файл только для чтения (`'r'` — _read_);
> * `encoding='UTF-8'` показывает, что файл по умолчанию прочитали в кодировке UTF-8.   

Последнее будет, если у вас MacOS или Linux. На Windows вы увидите `encoding='cp1251'`.
Отличия возникают, потому что у питона разное поведение по умолчанию в зависимости от операционной системы пользователя. 

````python
# Запущено на Windows
with open('first_text.txt') as infile:
    print(infile)  # Пробуем распечатать считанный файл
````

Ну и главное, что мы видим, это что если сделать print(<псевдоним>), содержимого мы не увидим. 

Чтобы уберечься и избежать умолчаний, мы можем явно написать, что файл хотим открывать для чтения `mode="r"`
    и что пробуем открыть в кодировке UTF-8 `encoding="utf-8"`.
    
И наконец, давайте же считаем файл.
Для этого у любого псевдонима, который мы запишем после `as`, будет метод `.read()`. 
**Метод `.read()` считывает файл целиком**.
Нужно быть осторожным, если файлы могут быть слишком большими.

````python
with open('first_text.txt', encoding='UTF-8') as f:
    text = f.read()
print(text)
````

Псевдоним файла это такая штучка, у которой есть метод `.read()` (а также некоторые другие), и по которой ещё можно проходиться в цикле!

Тогда мы будем получать отдельные строки файла на каждом шагу. 

````python
with open('first_text.txt', encoding='UTF-8') as f:
    for line in f:
        print(line)
        print('-----')
````

### Тренировка

Давайте отфильтруем строки файла так, чтобы получить в итоге лишь те, которые...
- длиннее 21
- содержат цифры

````python
# длиннее 21
strs_more_21 = []
with open('first_text.txt', encoding='UTF-8') as f:
    for line in f:
        if len(line) > 21:
            strs_more_21.append(line)

print(*strs_more_21)
````

````python
# содержат цифры
with open('first_text.txt', encoding='UTF-8') as f:
    for line in f:
        for symb in line:
            if symb.isdigit():
                print(line)
                break
````

## Открываем .csv

Как разумнее открывать таблицы?
    Сохранять строки в список.

````python
csv_data = []            
with open('gismeteo.csv', encoding='UTF-8') as infile: # открываем файл под псевдонимом
    for line in infile:  
        csv_data.append(line.strip('\n'))  
````

Чтобы работать со строкой выше (или списком строк как здесь) как с таблицей, всё же потребуются дополнительные действия, которые сейчас и попробуем выполнить.

### Тренировка
Создайте при чтении файла `gismeteo.csv` список списков:

```python
[['20230317', 4, -1, 6, 0.0], ['20230318', 3, -7, 5, 0.0], ...]
```

Каков план? Давайте сформулируем вместе.

````python
new_list = []
for line in csv_data[1:]:
    splitted_line = line.split(',')
    for i in range(len(splitted_line)):
        if i != 0:
            splitted_line[i] = float(splitted_line[i])
    new_list.append(splitted_line)
new_list    
````

- Давайте на основе этих списков посчитаем **среднюю температуру** для каждого .

````python
for line in new_list:
    aver = (line[1] + line[2])/2
    print(line[0], aver)
````

Для округления нам могут понадобиться функции:
* `int` (наша знакомая!) &mdash; просто отбросит дробную часть
* `ceil` (потолок) из `math` &mdash; приведёт строго вверх, к большему 
* `round` &mdash; сделает **банковское** округление  
(как математическое, если дробная часть меньшие половины, округлять вниз, а если дробная часть больше половины, округлять вверх),  
но $1,5 \rightarrow 2$, а $4,5 \rightarrow 4$ (при ,5 округляется к чётному)

````python
# округление вниз
print(int(4.4), int(4.5), int(4.6))
````

````python
# округление вверх
from math import ceil
print(ceil(4.4), ceil(4.5), ceil(4.6))
````

````python
# банковское округление
print(round(4.4), round(4.5), round(4.6))
print(round(3.4), round(3.5), round(3.6))
````

````python
# округлённая средняя темература
for line in new_list:
    aver = round((line[1] + line[2])/2)
    print(line[0], aver)
````

## Записываем в файл

Представим, что в строке `important_text` лежит какой-то важный текст, который нужно сохранить. 
    Снова воспользуемся конструкцией `with open() as`, но укажем, что открываем файл для записи: `'w'` — _write_.
    При необходимости можно дополнительно указать кодировку. 

Можно написать просто `'w'`, а не `mode='w'`, так как `open()` ожидает на втором месте именно этот аргумент.
    Если бы не указывали режим, а сразу передавали кодировку, то уже обязательно нужно было бы написать `encoding=`, так как по умолчанию она ожидается на третьем месте, а мы ее передаем на втором.

Печать в файл происходит с помощью функции `print()`, у которой в качестве аргумента указано, что надо печатать в файл с псевдонимом `outfile`. 

````python
important_text = 'Это очень важный текст. В нем даже несколько строк. \nОни отделены переносом строки.'

with open('info.txt', 'w', encoding='UTF-8') as outfile:    # Открываем файл на запись
    # Печатаем в файл, передавая его псевдоним в качестве аргумента функции print()
    print(important_text, file=outfile)
````

````python
important_text_2 = 'Это другой важный текст.'
with open('info.txt', 'w', encoding='UTF-8') as outfile:   # Открываем тот же файл
    print(important_text_2, file=outfile) 
````

Посмотрите, что лежит в файле `info.txt`.
    Там только содержимое `important_text_2`!
    Мы перезаписали файл.

А как дописать что-то в уже существующий файл?

- `mode='r'` — _read,  **чтение**_. При отсутствии указанного файла выдается ошибка. Это режим по умолчанию.
- `mode='w'` — _write, **запись**_. При отсутствии указанного файла создается новый. При наличии файла содержимое перезаписывается.
- `mode='a'` — _append, **дозапись**_. При отсутствии указанного файла создается новый. При наличии файла новое содержимое дозаписывается после имеющегося.

````python
important_text_3 = 'Это третий важный текст. Надеемся, что он запишется после второго'
with open('info.txt', 'a', encoding='UTF-8') as outfile:  # Открываем тот же файл, mode='a'
    print(important_text_3, file=outfile) 
````

````python
with open('info.txt', 'r', encoding='UTF-8') as f:  # Открываем тот же файл, mode='a'
    text = f.read()
print(text)
````
